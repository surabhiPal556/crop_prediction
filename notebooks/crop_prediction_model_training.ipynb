```python
##  Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##  Load Dataset
import pandas as pd
data=pd.read_csv('/content/crop.csv')
print(data)
```

```python
data.head()
```

```python
## Data Preprocessing
sns.heatmap(data.isnull())
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2, random_state=0)
```

Step 4: Model Training

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
Knn=KNeighborsClassifier(n_neighbors=5)
Knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
Knn_pred=Knn.predict(x_test)
```

```python
cmknn=confusion_matrix(y_test,Knn_pred)
sns.heatmap(cmknn,annot=True,fmt='d')
```

```python
crknn=classification_report(y_test,Knn_pred)
print(crknn)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logm=LogisticRegression()
logm.fit(x_train,y_train)
```

```python
logm_pred=logm.predict(x_test)
```

```python
cmlogm=confusion_matrix(y_test,logm_pred)
sns.heatmap(cmlogm,annot=True,fmt='d')
```

```python
crlogm=classification_report(y_test,logm_pred)
print(crlogm)
```

Comparative Analysis

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 1. Load Data (Ensuring the 'label' column is the target)
# data = pd.read_csv('crop.csv')
X = data.drop('label', axis=1)
y = data['label']

# 2. Split Data - Using random_state=0 to match your notebook's 99% accuracy
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# 3. Initialize Models
knn = KNeighborsClassifier(n_neighbors=5)
logm = LogisticRegression(max_iter=300)
nb = GaussianNB()
svm = SVC(kernel='linear') # Linear kernel often performs better for this dataset
dt = DecisionTreeClassifier(random_state=0)
rf = RandomForestClassifier(n_estimators=100, random_state=0)

# 4. Train and Calculate Accuracy
models = [knn, logm, nb, svm, dt, rf]
algorithms = ["KNN", "Logistic Regression", "Naive Bayes", "SVM", "Decision Tree", "Random Forest"]
accuracies = []

for model in models:
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, pred) * 100)

# 5. 🎨 VISUALIZATION (Matching teacher's color combination)
colors = ['orange', 'pink', 'skyblue', 'green', 'purple', 'red'] # Professional shades

plt.figure(figsize=(12, 7))
bars = plt.bar(algorithms, accuracies, color=colors, edgecolor='black', alpha=0.8)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5,
             f'{yval:.0f}', ha='center', va='bottom', fontweight='bold', fontsize=12)

# Formatting to match "Motor Fault" Slide 13
plt.title("COMPARATIVE ANALYSIS", fontsize=16, fontweight='bold', pad=20)
plt.ylabel("Accuracy (%)", fontsize=12, fontweight='bold')
plt.xlabel("Algorithms", fontsize=12, fontweight='bold')
plt.ylim(90, 100) # Focusing on the high-performance range
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
